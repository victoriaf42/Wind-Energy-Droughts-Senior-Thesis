```python
# Given values
capacity_mw = 100            # Wind farm size in MW
cost_per_kw = 1024           # Cost in $ per kW

# Convert MW to kW
capacity_kw = capacity_mw * 1000

# Calculate total cost
total_cost = capacity_kw * cost_per_kw

# Print result
print(f"Total installation cost for {capacity_mw} MW wind farm: ${total_cost:,.0f}")
```

```python
# ============================================
# Wind Farm Debt Service using CRF from WACC
# ============================================

# ---- Inputs ----
capacity_mw   = 100
cost_per_kw   = 1024          # $/kW (CapEx)
wacc_nominal  = 0.0625        # 6.25% nominal
life_years    = 25
inflation     = 0.025         # 2.5%

# ---- CapEx ----
capex = capacity_mw * 1000 * cost_per_kw  # MW -> kW -> $

# ---- Capital Recovery Factor (CRF) from WACC ----
r = wacc_nominal
n = life_years
crf = (r * (1 + r)**n) / ((1 + r)**n - 1)

# ---- Annual / Monthly payments (Nominal) ----
annual_payment_nominal  = capex * crf
monthly_payment_nominal = annual_payment_nominal / 12

# ---- "Real" (inflation-adjusted) equivalents in year-0 dollars ----
annual_payment_real  = annual_payment_nominal / (1 + inflation)
monthly_payment_real = annual_payment_real / 12

# ---- Output ----
print("===== Wind Farm Debt Service (CRF computed from WACC) =====")
print(f"CapEx: ${capex:,.0f}")
print(f"Nominal WACC (r): {r:.4%}")
print(f"Life (n): {n} years")
print(f"Computed CRF: {crf:.4%}")

print("\n--- Nominal levelized payments ---")
print(f"Annual:  ${annual_payment_nominal:,.0f}")
print(f"Monthly: ${monthly_payment_nominal:,.0f}")

print("\n--- Inflation-adjusted (approx. year-0 $) ---")
print(f"Annual (real):  ${annual_payment_real:,.0f}")
print(f"Monthly (real): ${monthly_payment_real:,.0f}")
```
